```julia
using Plots
```

```julia
models = ["JuloVelo", "scVelo", "DeepVelo", "cellDancer"]
times = [9.28689, 3.95348, 7.54070, 28.87956]
```

```julia
theme(:vibrant, framestyle = :axes, grid = false, markersize = 3, linewidth = 0, palette = :tab20) 
```

```julia
p = bar(models, times, label = nothing, tickfontsize = 12, guidefontsize = 14, bar_width = 0.3, ylabel = "Execution time (Minutes)", color = "gray", xlabel = "Models", dpi = 300)
```

```julia
savefig(p, "Execute_time.png")
```
